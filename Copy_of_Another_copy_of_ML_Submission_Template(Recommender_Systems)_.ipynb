# **Project Summary -**

Recommender systems have gained tremendous populatity of late owing to the rise of digital world. From social media to e-commerce, the applications of recommendation system is increasing at a rapid pace.
In line with its rapid use,this project too aims to develop a book recommendation system. We have 3 datasets - Users,Books and Ratings. We undertake the following steps to perform the required task.

1. Quick understanding of datasets, by **loading them,getting information,description,understanding features, missing values,unique values etc.**
2. Next, we perform **Exploratory Data Analysis(EDA)**, beggining with data wrangling operations to convert data to usable format.We delete missing values rows, infil the missing values and outliers according to the situation. We dropped the unnecessary columns from the dataframes. Merged dataframes to get all encompassing  dataframe, to carry on further operations.
3. We carry out necessary datatype conversions, and further proceed to get subdataframes displaying information like most read books,most rated books, users rated most books etc.
4. Then we proceed to **(Data Visualization)**, In it we perform univariate,bivariate and multivariate analysis keeping Book-Rating as the target variable.
5. Thereafter we perform **feature engineering and Data Pre-processing.**Textual features are dealt accordingly as in **Natural Language Processing(NLP) like-Expanding contractions, Punctuation removing, Removing URLs, Removing words containing digits, Converting to lower case, Removing white spaces,Tokenization, Lemmatization and Vectorization.** New features are also created from existing like Country and Age-Group.
6. Thereafter, we proceed to implement various **algorithms** to recommend Books to the users.
7. **Content filtering based recommendor system.**
This model recommends that books to users, which are similar to the book the user has already ineracted with.For its implementation, we select those 25000 books, which have received most ratings as they are most popoular among users.A TF-IDF matrix is created based on features of all the books like-Title,Author,Publisher and Year of Publication.This matrix is used to generate the cosine similarity marix.This similarity matrix records the extent of similarity between different pairs of all the books.Next, we define a function, which takes the title of interacted book and cosine similarity matrix as input.We consider the similarity scores of rest of the books, corresponding to the interacted book.We get the top n scores and get ISBN and Book Titles for those top n scores.Thus we implement a Content based book recommendation system.
8. **User based collaborative recommender system.**
This algorithm utilizes the interactions made by similar users to recommend books for a particular user.These interactions are based on ratings given to various books by the users. To illustrate, we take a sample dataframe with users who have rated most books. We utilize pivot_table function to create user item matrix from this dataframe.This matrix is used to create similarity matrix.This similarity matrix shows the similarity coefiicients between various users. Next we define a function, which takes user-id and n as input,to return book suggestions. In this function, we get the similiarity scores for the target user. Then we iterate over each book in user item matrix, skipping the ones already rated by the target user.  We calculate the weighted sum of ratings for each book based on the user similarity. This is done by iterating over each user and multiplying their similarity score with the target user by their rating for the book. Thereafter, we calculate the average similarity score for the current book. This is done by summing the similarity scores of all users who have rated the book and dividing by the count of such users. We calculate the predicted rating for each book ,append them to list, sort it and return the top n books.Thus we implement a Collaborative recommendation system.
9. **Demographic based recommender system.**
It uses the demographic information of users, such as age, gender, location, etc., to generate recommendations for users.  This model intends to solve 'COLD START' problem, by recommending books to users with no prior history of interaction. We define a function which takes user_id,age and country as input arguments and returns top n books.The books are filtered based on age, followed by location.Next, we remove null values and deduplicate the result.
Lastly, books are sorted in descending order of ratings and recommended.
10. At last, we intend to combine all three systems as future work to improve recommendations.
Demographic based algorithm can work well for new users avoiding cold start problem.Content based and collborative systems can be  combined in hybrid recommender systems to improve the quality of recommendations.

### ML Model - 2 User based Collaborative filtering

**This algorithm utilizes the interactions made by similar users to recommend books for a particular user.These interactions are based on ratings given to various books by the users.It further improves the recommendations made.**

````python
# ML Model - 2 Implementation
# Here a sample dataframe(25 rows) is considered to reduce execution time, given the resource constraints.
# We shall take top 25 users which have given most ratings.
df_ratings_orig=df_ratings
y=bri_desc.iloc[0:3]
y=y['User-ID'].drop_duplicates()
y=y.astype(int)
y=y.astype(str)
df_ratings  = df_ratings[df_ratings['User-ID'].isin(list(y))]
````

````python
# Create a user-item rating matrix
user_item_matrix = pd.pivot_table(df_ratings, values='Book-Rating', index='User-ID', columns='ISBN', fill_value=0)
````

````python
user_item_matrix
````

````python
# Calculate the cosine similarity between users
from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(user_item_matrix)
user_similarity = pd.DataFrame(user_similarity,index=list(user_item_matrix.index),columns=[user_item_matrix.index])
user_similarity
````

````python
# Define a function to get book recommendations for a user
def get_recommendations(user_id, top_n=5):
    # Get the similarity scores for the target user
    target_user_similarity = user_similarity.loc[(user_id)]

    # Get the user's ratings
    user_ratings = user_item_matrix.loc[user_id]

    # Initialize an empty list to store recommendations
    recommendations = []

    # Iterate over each book
    for book_id in user_item_matrix.columns:
        # Skip books the user has already rated
        if user_ratings[book_id] > 0:
            continue

        # Calculate the weighted sum of ratings for this book based on user similarity
        # This is done by iterating over each user and multiplying their similarity score with the target user by their rating for the book.
        weighted_sum = sum(target_user_similarity.loc[other_user_id] * user_item_matrix.loc[other_user_id][book_id]
                           for other_user_id in list(user_item_matrix.index))/len(target_user_similarity)

        # Calculate the average similarity score for the current book
        # This is done by summing the similarity scores of all users who have rated the book and dividing by the count of such users.
        rated_users_similarity_sum = sum(target_user_similarity.loc[other_user_id]
                                         for other_user_id in user_item_matrix.index  if user_item_matrix.loc[other_user_id, book_id] > 0)

        try:
          average_similarity = rated_users_similarity_sum / len([other_user_id for other_user_id in user_item_matrix.index
                                                                if user_item_matrix.loc[other_user_id][book_id] > 0])

        except  ZeroDivisionError:
          average_similarity = 0

        # Calculate the predicted rating for this book
        if ((average_similarity > 0) is True):
          predicted_rating = weighted_sum / average_similarity
        else:
          predicted_rating  = 0

        # Add the book ID and predicted rating to the recommendations list
        recommendations.append((book_id, predicted_rating))

    # Sort recommendations by predicted rating in descending order
    recommendations.sort(key=lambda x: x[1], reverse=True)

    # Get the top N recommended book IDs
    top_recommendations = [book_id for book_id, _ in recommendations[:top_n]]
    return top_recommendations
````

#### 1. Explain the ML Model used in this recommendation system.

**User based collaborative recommender system.**
This algorithm utilizes the interactions made by similar users to recommend books for a particular user.These interactions are based on ratings given to various books by the users. To illustrate, we take a sample dataframe with users who have rated most books. We utilize pivot_table function to create user item matrix from this dataframe.This matrix is used to create similarity matrix.This similarity matrix shows the similarity coefiicients between various users. Next we define a function, which takes user-id and n as input,to return book suggestions. In this function, we get the similiarity scores for the target user. Then we iterate over each book in user item matrix, skipping the ones already rated by the target user.  We calculate the weighted sum of ratings for each book based on the user similarity. This is done by iterating over each user and multiplying their similarity score with the target user by their rating for the book. Thereafter, we calculate the average similarity score for the current book. This is done by summing the similarity scores of all users who have rated the book and dividing by the count of such users. We calculate the predicted rating for each book ,append them to list, sort it and return the top n books.Thus we implement a Collaborative recommendation system.

````python
# Define a function to recommend top-rated books at a given location
def recommend_books_by_location(user_id, age, country, top_n=5):

    # Filter books by age_group
    age_group = df_total[df_total['Age'] == age].iloc[0]['Age-Group']
    age_books = df_total[df_total['Age-Group'] == age_group]

    # Further Filter books of the specified location
    location_books = age_books[age_books['Country'] == location]

    # Exclude books the user has already rated
    user_rated_books = df_ratings[df_ratings['User-ID'] == user_id]['ISBN']
    location_books = location_books[~location_books['ISBN'].isin(user_rated_books)]

    # Remove Nan values
    location_books  = location_books[location_books.notna()]

    #Deduplicate
    location_books  = location_books.drop_duplicates()

    # Sort books by average rating in descending order
    location_books = location_books.sort_values('Book-Rating', ascending=False)

    # Get the top N recommended books ISBN
    top_recommendations = list(location_books['Book-Title'].head(top_n))
    return top_recommendations
````
